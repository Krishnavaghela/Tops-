```python
# 1. Compute the minimum, 25th percentile, median, 75th, and maximum of ser.

# Create a random series with 20 values from a normal distribution
num_state = np.random.RandomState(100)
num_series = pd.Series(num_state.normal(10, 4, 20))

# Display the original series
print("Original Series:")
print(num_series)

# Calculate the minimum, 25th percentile, median, 75th percentile, and maximum
result = np.percentile(num_series, q=[0, 25, 50, 75, 100])

# Print the computed values
print("\nMinimum, 25th percentile, median, 75th, and maximum of the given series:")
print(result)
```

```python
# 9. Create a line plot using the x and y values provided below. Label the y-axis as “Y” and label the x-axis as “X”.
x = [3, 4, 5, 6]
y = [1.5, 2, 2.5, 3]

plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot')
plt.grid(True)
plt.show()
```

```python
# 10. Create an array of numbers between 0 and 6 with increments of 0.3 and name its “x”. Then on the same plot, plot x, x², x³, and x⁴. For consistency, use the following style lines repectvely “ro”, “bs”, “g”, and “:”. Lastly, make sure that the x-axis covers 0 to 6, while the y-axis spans from 0 to 125. Do not worry if you are not familiar with the style lines — you will recognize them as soon as you see the plot.

x = np.arange(0, 6.3, 0.3)

x_squared = x**2
x_cubed = x**3
x_quartic = x**4

plt.figure(figsize=(8, 6))

plt.plot(x, x, 'ro', label='x')
plt.plot(x, x_squared, 'bs', label='x²')
plt.plot(x, x_cubed, 'g', label='x³')
plt.plot(x, x_quartic, ':', label='x⁴')

plt.xlim(0, 6)
plt.ylim(0, 125)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x, x², x³, and x⁴')
plt.legend()

plt.show()
```

```python
# 11. Heights and initials of a group of individuals are provided below. Create a bar plot titled “Height Comparison” to compare the heights among this group.
height = [179, 155, 191, 152, 188, 177]
names = ['QA', 'WB', 'EC', 'RD', 'TE', 'YF']

plt.figure(figsize=(8, 6))
plt.bar(names, height, color='skyblue')
plt.xlabel('Names')
plt.ylabel('Height (cm)')
plt.title('Height Comparison')
plt.grid(axis='y')
plt.show()
```

```python
# 12. Plot a histogram of x, where x consists of 100,000 randomly-selected points with a normal distribution (hint: you can use numpy.random.randn() to generate the random points). The histogram should have 10 bins. Look at how the histogram changes when we try 20 and 50 bins.

x = np.random.randn(100000)

# Create a histogram with 10 bins
plt.figure(figsize=(10, 6))
plt.hist(x, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram with 10 Bins")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Create a histogram with 20 bins
plt.figure(figsize=(10, 6))
plt.hist(x, bins=20, color='salmon', edgecolor='black')
plt.title("Histogram with 20 Bins")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Create a histogram with 50 bins
plt.figure(figsize=(10, 6))
plt.hist(x, bins=50, color='green', edgecolor='black')
plt.title("Histogram with 50 Bins")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
